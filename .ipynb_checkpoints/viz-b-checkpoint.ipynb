# Seaborn Visualization


## importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## setting seaborn as default plotter

```python
sns.set_theme(style="dark")
```

## getting datasets available by seaborn

```python
sns.get_dataset_names()
```

## loading wanted dataset, diamonds in our case

```python
penguins = sns.load_dataset('penguins')
data = pd.DataFrame(data=penguins)
data.head(10)
```

## plotting the relation between the bill depth in MM and body mass in Gram 

```python
sns.relplot(x="bill_depth_mm", y="body_mass_g", hue="species", style="sex", data=data)
```

## plotting the relation between sex and body mass in Gram

```python
sns.relplot(x="sex", y="body_mass_g", hue="species", data=data)
```

## Conclusions
* the relation between the bill depth and the body mass as noted above is negative relation
* gentoo species are more likely to have higher body mass
* males in general are the sex with highest body mass whatever the species were
