# Seaborn Visualization


## importing required libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## setting seaborn as default plotter

```python
sns.set()
```

## getting datasets available by seaborn

```python
sns.get_dataset_names()
```

## loading wanted dataset, tips in our case

```python
tips = sns.load_dataset('tips')
data = pd.DataFrame(data=tips)
data.head(10)
```

## plotting the relation between the tip and total bill

```python
sns.relplot(x="total_bill", y="tip", data=data);
```

## plotting again after adding other factor, sex, to study the relation between these 3 factors

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="sex", data=data);
```

## plotting the relation between the tip and the day and paying attention to somker/non smoker

```python
sns.relplot(x="day", y="tip", hue="smoker", data=data)
```

```python
len(data[data["smoker"] == "Yes"])
```

```python
len(data[data["smoker"] == "No"])
```

## plotting the relation between the total bill and the day and paying attention to somker/non smoker

```python
sns.relplot(x="day", y="total_bill", hue="smoker", data=data)
```

## Conclusions
* the relation between the total bill and the tip as noted above is positive relation
* males are more likely to leave higher tip
* non smokers are most people to leave a tip
* saturday is the day with highest rate of bills, therefore, it has the highest rate of tips
